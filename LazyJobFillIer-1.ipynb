```python
import requests
import re
from bs4 import BeautifulSoup
```

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
```

```python
def fill_df(df):
    for i in range(len(df)):
        if type(df.iloc[i][0]) == float and type(df.iloc[i]["Link"]) == str and "linkedin" in df.iloc[i]["Link"]:
            link = df.iloc[i]["Link"]
            response = requests.get(link)
            soup = BeautifulSoup(response.text)
            # get job title
            job_title = soup.find_all('h1')
            job_title = re.sub("<(.)+?>", "", str(job_title))[1:-1]
            # get company name
            company_name = soup.find_all('a')
            company_name = [str(c) for c in company_name]
            company_name = [c for c in company_name if "topcard__org-name-link topcard" in c]
            if company_name:
                company_name = company_name[0]
                company_name = re.sub("<(.)+?>", "", company_name)
                company_name = re.sub("( |\\n)", "", company_name)
                # fill it in
                df.iloc[i][0] = job_title
                df.iloc[i][1] = company_name
    return df
```

```python
df = fill_df(df)
```

```python
with open("job_apps.txt", "w") as f:
    f.write(",".join(list(df.columns)) + "\n")
    for i in range(len(df)):
        f.write(",".join([str(c) for c in list(df.iloc[i])])+"\n")
    
```
