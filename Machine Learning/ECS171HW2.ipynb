## Exercise 1:

```python
import pandas as pd
import numpy as np
```

```python
data_pre = pd.read_csv("garments_worker_productivity.csv")
data_pre.head()
```

#### 1.1: Dropping the date

```python
data_pre = data_pre.drop(["date"], axis = 1)
data_pre.head()
```

#### 1.2: Printing out unique categories for categorical variables

```python
# list of categorical variables;
cat_atts = ["day", "quarter", "department", "team"]

for cat in cat_atts:
    print(cat, data_pre[cat].unique())
```

#### 1.3: Remapping the duplicates/typos in categorical data:

```python
data_pre["department"] = data_pre["department"].replace("sweing","sewing")
data_pre["department"] = data_pre["department"].replace("finishing ","finishing")

print(data_pre["department"].unique())
```

#### 1.4: Creating "satisfied" column

```python
satisfaction = []

for i in range(0, len(data_pre)):
    if data_pre["actual_productivity"][i] >= data_pre["targeted_productivity"][i]:
        satisfaction.append(1)
    else:
        satisfaction.append(0)

data_pre["satisfied"] = satisfaction
data_pre.head()
```

#### 1.5: Dropping columns "actual_productivity" and "targeted_productivity"

```python
data_pre = data_pre.drop(["actual_productivity", "targeted_productivity"], axis = 1)
data_pre.head()
```

#### 1.6: Find, print, and replace NA, NaN, null, None

```python
for col in data_pre:
    print(col, ":", any(data_pre[col].isna()))
```

#### "wip" is the only column with NaNs

```python
data_pre = data_pre.fillna(0)
```

```python
df = data_pre
df.head()
```

```python
import sklearn
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
```

#### 2.2: Additional Data Preprocessing

#### 2.2: Naïve Bayes Classifier for Categorical Attributes

```python
from sklearn.naive_bayes import CategoricalNB
from sklearn.metrics import classification_report
```

```python
cat_atts = ["day", "quarter", "department", "team"]

catNB = CategoricalNB()
```

```python
train, test = train_test_split(data_pre, test_size=0.2, random_state=21)
X_train, y_train = train[cat_atts], train["satisfied"]
X_test, y_test = test[cat_atts], test["satisfied"]
y_train.value_counts()
```

#### 3.3: SVM with Over-sampling

```python
from sklearn.utils import resample
```

```python
# printing out the frequency of each class

print(df3["satisfied"].value_counts())
```

```python
sat = df3[df3['satisfied'] == 1]
notsat = df3[df3['satisfied'] == -1]

resampled_notsat = resample(notsat, n_samples = 875, random_state=21)

df4 = sat.append(resampled_notsat)
df4["satisfied"].value_counts()
```
